```python
import pandas as pd
import re
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("house_prices.csv")
df.shape
```

```python
df.head(20)
```

```python
df.drop(['Index', 'Description', 'Society'], inplace=True, axis=1)
```

```python
df.isna().sum()
```

```python
df.drop(['Plot Area', 'Dimensions'], inplace=True, axis=1)
df.dropna(subset=['Bathroom','Floor', 'Price (in rupees)'], inplace=True)
```

```python
df.head(20)
```

```python
df['Status'].value_counts()
```

```python
df.drop(['Status', 'Transaction', 'Amount(in rupees)'], inplace=True, axis=1)
```

```python
df.isna().sum()
```

```python
df['Balcony'] = df['Balcony'].fillna(0)
df['Car Parking'] = df['Car Parking'].fillna(0)
df['Bathroom'] = df['Bathroom'].fillna(0)
```

```python
df.isna().sum()
```

```python
df.head(20)
```

```python
a=["Carpet Area", "location","Floor","Furnishing","facing","overlooking","Bathroom","Balcony","Car Parking","Ownership"]
for cat in a :
    print(f"{cat}:\n{df[cat].unique()}\n")
```

```python
def clean_area(col):
    return pd.to_numeric(col.astype(str).str.replace("sqft", "", regex=False).str.strip(), errors='coerce')

df['Carpet Area'] = clean_area(df['Carpet Area'])
df['Super Area'] = clean_area(df['Super Area'])
```

```python
for col in ['Bathroom', 'Balcony']:
    df[col] = df[col].astype(str).replace(['> 10', '>10'], '11')
    df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)
```

```python
df.head(20)
```

```python
def convert_parking(value):
    if pd.isna(value):
        return 0
    value_str = str(value)
    num_match = re.search(r'(\d+)', value_str)
    number = float(num_match.group(1)) if num_match else 0
    return number * 1.3 if 'Covered' in value_str else number

df['Car Parking'] = df['Car Parking'].apply(convert_parking)
```

```python
def extract_floor_info(value):
    if pd.isna(value):
        return (np.nan, np.nan)
    value = str(value)
    match = re.search(r'(\d+)\s*out of\s*(\d+)', value)
    if match:
        return (int(match.group(1)), int(match.group(2)))
    elif 'Ground' in value:
        match2 = re.search(r'out of\s*(\d+)', value)
        return (0, int(match2.group(1)) if match2 else np.nan)
    elif 'Upper Basement' in value:
        match2 = re.search(r'out of\s*(\d+)', value)
        return (-1, int(match2.group(1)) if match2 else np.nan)
    elif 'Lower Basement' in value:
        match2 = re.search(r'out of\s*(\d+)', value)
        return (-2, int(match2.group(1)) if match2 else np.nan)
    else:
        return (np.nan, np.nan)

df[['CurrentFloor', 'TotalFloors']] = df['Floor'].apply(lambda x: pd.Series(extract_floor_info(x)))
df.drop(columns=['Floor'], inplace=True)
```

```python
a=["Furnishing","Super Area","Bathroom","Balcony","Car Parking", "CurrentFloor", "TotalFloors"]
for cat in a :
    print(f"{cat}:\n{df[cat].unique()}\n")
```

```python
mapping = {
    'Unfurnished': 0,
    'Semi-Furnished': 1,
    'Furnished': 2
}

df['Furnishing'] = df['Furnishing'].fillna('Unfurnished').map(mapping)
```

```python
df['overlooking'] = df['overlooking'].fillna('Not Available')
df = df.join(df['overlooking'].str.get_dummies(sep=', ').add_prefix('overlooking_'))
df.drop(columns=['overlooking'], inplace=True)
```

```python
df['facing'] = df['facing'].fillna('Not Available')
df = df.join(df['facing'].str.get_dummies(sep=', ').add_prefix('facing_'))
df.drop(columns=['facing'], inplace=True)
```

```python
df['Ownership'] = df['Ownership'].fillna('Not Available')
df = df.join(df['Ownership'].str.get_dummies(sep=', ').add_prefix('Ownership_'))
df.drop(columns=['Ownership'], inplace=True)
```

```python
df['Bedroom'] = df['Title'].str.extract(r'(\d+)\s*BHK', expand=False)
df['Bedroom'] = pd.to_numeric(df['Bedroom'], errors='coerce')
df.drop(['Title'], inplace=True, axis=1)
```

```python
loc_freq = df['location'].value_counts(normalize=True)
df['location_encoded'] = df['location'].map(loc_freq)
df.drop(columns=['location'], inplace=True)
```

```python
df['Super Area'] = df['Super Area'].fillna(0)
```

```python
df.head(20)
```

```python
a=["Super Area", "location_encoded", "Bedroom"]
for cat in a :
    print(f"{cat}:\n{df[cat].unique()}\n")
```

```python
df['CurrentFloor'] = df['CurrentFloor'].fillna(0)
df['TotalFloors'] = df['TotalFloors'].fillna(0)
df['Bedroom'] = df['Bedroom'].fillna(0)
```

```python
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

mean_carpet_per_bedroom = train_df.groupby('Bedroom')['Carpet Area'].mean()

train_df['Carpet Area'] = train_df.apply(
    lambda row: mean_carpet_per_bedroom[row['Bedroom']] if pd.isna(row['Carpet Area']) else row['Carpet Area'],
    axis=1
)

test_df['Carpet Area'] = test_df.apply(
    lambda row: mean_carpet_per_bedroom[row['Bedroom']] if pd.isna(row['Carpet Area']) else row['Carpet Area'],
    axis=1
)
```

```python
train_df['Carpet Area'] = df['Carpet Area'].fillna(0)
test_df['Carpet Area'] = df['Carpet Area'].fillna(0)
```
